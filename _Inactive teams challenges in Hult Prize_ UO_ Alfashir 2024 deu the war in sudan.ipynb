### Outlined below are the highlighted topics in the notes.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer_notes = TfidfVectorizer(max_features=1000, stop_words='english')
X_notes = vectorizer_notes.fit_transform(datas['notes'].fillna(''))  

feature_names_notes = vectorizer_notes.get_feature_names_out()
tfidf_scores_notes = X_notes.sum(axis=0).A1
tfidf_df_notes = pd.DataFrame({'word': feature_names_notes, 'score': tfidf_scores_notes}).sort_values(by='score', ascending=False)

print("The most influential words in the notes:\n", tfidf_df_notes.head(20))
```

#### It appears that the network issue related to recording lectures is  the most frequently occurring topic, in addition to the time difference, which has shown that timing is one of the subproblems during the analysis phase.
